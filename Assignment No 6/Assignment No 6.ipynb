2. Import Libraries

```python
from transformers import MarianMTModel, MarianTokenizer
```

3. Load Pre-trained Translation Model

```python
model_name = "Helsinki-NLP/opus-mt-en-hi"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

4. Translation Function

```python
def translate_text(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    output = tokenizer.decode(translated[0], skip_special_tokens=True)
    return output
```

5. Translate Public Information Content

```python
english_text = """
The government has launched a new digital health program 
to provide free medical services to citizens.
"""

translated_text = translate_text(english_text)
print("English Text:\n", english_text)
print("\nTranslated Text (Hindi):\n", translated_text)
```

```python
6. Translating Multiple Sentences
```

```python
sentences = [
    "Public transport services will remain closed tomorrow.",
    "Citizens are advised to follow safety guidelines."
]

for s in sentences:
    print("EN:", s)
    print("HI:", translate_text(s))
    print()
```

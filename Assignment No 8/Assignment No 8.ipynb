```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

# Download required datasets (run only first time)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

2. Example of Ambiguous Sentence

```python
sentence = "He went to the bank to deposit money"
words = word_tokenize(sentence)

# Apply Lesk algorithm
sense = lesk(words, "bank")

print("Sentence:", sentence)
print("Predicted Sense:", sense)
print("Definition:", sense.definition())
```

3. Another Example

```python
sentence2 = "The river overflowed the bank after heavy rain"
words2 = word_tokenize(sentence2)

sense2 = lesk(words2, "bank")

print("\nSentence:", sentence2)
print("Predicted Sense:", sense2)
print("Definition:", sense2.definition())
```

4. Check Synonyms, Hypernyms, Antonyms

```python
if sense:
    print("\nSynonyms:")
    for lemma in sense.lemmas():
        print("-", lemma.name())
        
    print("\nHypernyms:")
    for hyper in sense.hypernyms():
        print("-", hyper.name())
        
    print("\nAntonyms:")
    for lemma in sense.lemmas():
        if lemma.antonyms():
            print("-", lemma.antonyms()[0].name())
```
